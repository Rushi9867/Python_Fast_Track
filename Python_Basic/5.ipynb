```python
# Scatterplot
sns.scatterplot(x="tip",y="total_bill",data=tips,hue="day",size="size",palette="rainbow_r")
```

```python
# Histogram
sns.histplot(tips['tip'],kde=True,bins=15)
```

```python
# Distplot
sns.distplot(tips['tip'],kde=True,bins=15)
```

```python
# Displot same as Hisplot
sns.displot(tips['tip'],kde=True,bins=15)
```

```python
# Barplot
sns.barplot(x="sex",y="tip",data=tips,palette="gist_rainbow_r")
```

```python
# Boxplot
sns.boxplot(x='day',y='tip',data=tips,hue='sex',palette='Purples_r')
```

```python
# Stripplot
sns.stripplot(x='day',y='tip',data=tips,hue='sex',palette='plasma',dodge=True)
```

```python
# Jointplot
sns.jointplot(x="tip",y="total_bill",data=tips,kind="kde",shade=True,cmap="ocean_r")# kind = reg
```

```python
sns.jointplot(x="tip",y="total_bill",data=tips,kind="hex",cmap="nipy_spectral")
```
